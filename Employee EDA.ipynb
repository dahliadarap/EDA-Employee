Exploratory Data Analysis

```python
#The Most Employee's Education Title
sns.countplot(x='Education', data = df, palette='Set1')
```

```python
#Turnover VS Education
sns.countplot(x='Education', hue = 'LeaveOrNot', data = df, palette='Set1')
```

```python
#Total of Employees that Joining per Year
sns.histplot(x='JoiningYear',data=df, kde=True, color='Green')
```

```python
#Joining Year VS City 
sns.countplot(x='JoiningYear', hue ='City', data = df, palette='Set2')
```

```python
#Turnover VS City
sns.countplot(x='City', hue ='LeaveOrNot', data = df, palette='Set2')
```

```python
#Turnover VS Payment Tier
sns.countplot(x='PaymentTier', hue ='LeaveOrNot', data = df, palette='Set3')
```

```python
#Turnover VS Age
sns.countplot(x='Age', hue ='LeaveOrNot', data = df, palette='Set1')
```

```python
#Total of Employees by Gender
data_gender = df['Gender'].value_counts()
ax = data_gender.plot(
    kind ='pie',
    autopct ='%1.0f%%',
    legend= True
)
```

```python
#Turnover VS Gender
sns.countplot(x='Gender', hue ='LeaveOrNot', data = df, palette='Set2')
```

```python
#Turnover VS Ever Benched
sns.countplot(x='EverBenched', hue ='LeaveOrNot', data = df, palette='Set3')
```

```python
#Turnover VS Experience In Current Domain
sns.countplot(x='ExperienceInCurrentDomain', hue ='LeaveOrNot', data = df, palette='Set1')
```
